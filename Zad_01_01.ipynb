#Подключение к базам данных

Одной из самых распространенных вещей, которые вам нужно делать в биоинформатике, является извлечение информации из биологических баз данных. Доступ к этим базам данных вручную может быть довольно утомительным, особенно если у вас много повторяющейся работы. Biopython пытается сэкономить ваше время и энергию, делая некоторые онлайновые базы данных доступными из скриптов Python. В настоящее время в Biopython есть код для извлечения информации из следующих баз данных:
* Entrez (https://www.ncbi.nlm.nih.gov/Web/Search/entrezfs.html) (and PubMed(https://www.ncbi.nlm.nih.gov/PubMed/)) from the NCBI – See Chapter ‍9.
* ExPASy(https://www.expasy.org/) – See Chapter ‍10.
* SCOP(http://scop.mrc-lmb.cam.ac.uk/scop/) – See the Bio.SCOP.search() function.

Код в этих модулях в основном упрощает написание кода Python, взаимодействующего со сценариями CGI на этих страницах, чтобы вы могли получать результаты в удобном для работы формате. В некоторых случаях результаты могут быть тесно интегрированы с парсерами Biopython, что еще больше упрощает извлечение информации.

#Глава 3. Объекты последовательностей

Биологические последовательности, возможно, являются центральным объектом биоинформатики, и в этой главе мы представим механизм Biopython для работы с последовательностями, объект Seq. В главе  ‍4 будет представлен связанный объект SeqRecord, который объединяет информацию о последовательности с любой аннотацией, снова используемой в главе  ‍5 для ввода последовательности.

Последовательности, по сути, представляют собой цепочки букв, такие как AGTACACTGGT, что кажется очень естественным, поскольку это наиболее распространенный способ просмотра последовательностей в биологических форматах файлов.

Самое важное различие между объектами Seq и стандартными строками Python заключается в том, что они имеют разные методы. Хотя объект Seq поддерживает многие из тех же методов, что и обычная строка, его метод translate() отличается тем, что выполняет биологический перевод, а также существуют дополнительные биологически важные методы, такие как reverse_complement().

```python
# Последовательности действуют как строки
for index, letter in enumerate(my_seq):
  print("%i %s" % (index, letter))
```

```python
# Вы можете получить доступ к элементам последовательности так же, как и к строкам (но помните, Python считает с нуля!):
print(my_seq[0])  # first letter
print(my_seq[2])  # third letter
print(my_seq[-1])  # last letter
```

```python
# Объект Seq имеет метод .count(), как и строка. Обратите внимание, что это означает, что, как и строка Python, это дает неперекрывающийся счет:
# Метод count() возвращает количество раз, когда указанный элемент появляется в списке.
from Bio.Seq import Seq
print("AAAA".count("AA"))
Seq("AAAA").count("AA")
```

```python
# Для некоторых биологических целей вам может потребоваться перекрывающийся счет (например, 3 в этом тривиальном примере). При поиске отдельных букв это не имеет значения:
my_seq = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
len(my_seq)
```

```python
my_seq.count("G")
```

```python
100 * (my_seq.count("G") + my_seq.count("C")) / len(my_seq)
```

Хотя вы можете использовать приведенный выше фрагмент кода для расчета GC%, обратите внимание, что модуль Bio.SeqUtils имеет несколько встроенных функций GC. Например:

```python
from Bio.SeqUtils import gc_fraction
gc_fraction(my_seq)
# Обратите внимание, что использование функции Bio.SeqUtils.gc_fraction() должно автоматически справляться
#  с последовательностями смешанного регистра и неоднозначным нуклеотидом S, что означает G или C.
```
